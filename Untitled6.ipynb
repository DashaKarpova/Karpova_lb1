```python
import collections
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sb
from scipy.stats import norm
from scipy.stats import chi2
```

```python
x=[20]
m=2**14
a=19
i=0
sol=[x[0]/m]
while (len(sol)!=1000):
  x.append((x[-1]*a)%m)
  i=x[-1]/m
  sol.append(i)
n=len(sol)
print(len(set(sol)))
```

```python
data=sol
sb.displot(data, stat='density')
```

```python
sb.displot(data, kind='ecdf')
```

```python
t=1
av=np.mean(sol)
nminus=1
nplus=0
alpha=0.05
for i in range(1,n):
  if sol[i]-av<0:
    nminus+=1
  else:
    nplus+=1
  if(sol[i]-av)*(sol[i-1]-av)<0:
    t+=1
m=((2*nplus*nminus)/n)+1
s=((m-1)*(m-2))/(n-1)
z_a=norm.ppf(1-alpha)
z=(t-m)/np.sqrt(s)
print(z_a)
print(abs(z))
print(t)
print(m-z_a*s**0.5,m+z_a*s**0.5)
```

```python
sol=sorted(sol)
xmin=min(sol)
xmax=max(sol)
degrees=(11)-1
critical=chi2.ppf(1-alpha,degrees)
h=1/11
v=[]
hp=h
chi_2=0
hi=0
for i in sol:
  v.append(sol.count(i))
i=0
while int(hp)<=1:
  if ((i<1000) and (sol[i]<hp)):
    hi+=v[i]
    i+=1
  else:
    hp+=h
    chi_2+=hi**2
    hi=0
print((chi_2*11/n-n),critical)
```
